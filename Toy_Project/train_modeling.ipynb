### 설명적 회귀 분석

```python
import statsmodels.api as sm

X = scaled_variable_frame
y = df['인원수']

# 상수항 추가
X = sm.add_constant(X)

# 회귀 모델 적합
model = sm.OLS(y, X).fit()
```

```python
print(model.summary())

# R-squared : 결정계수, 전체 데이터 중 해당 회귀모델이 설명할 수 있는 데이터의 비율, 회귀식의 설명력을 나타낸다성능을 나타냄(1에 가까울수록 성능이 좋음)
# Adj. R-squared: 모델에 도움이 되는 데이터에 따라 조정된 결정계수. 위 모델은 해당 회귀모델은 데이터의 46.4%을 설명한다.
# F-statistics: F 통계량으로 도출된 회귀식이 적절한지 볼 수 있음, 0과 가까울 수록 적절한 것
# Prob (F-statistic): 회귀식이 유의미한지 판단. (0.05이하일 경우 변수 끼리 매우 관련있다고 판단)
# std err: 계수 추정치의 표준오차(standard error), 값이 작을 수록 좋음
# t: t-test, 독립변수와 종속변수 사이의 상관관계, 값이 클 수록 상관도가 큼
# p-value(P > |t|): 독립변수들의 유의 확률, 0.05보다 작아야 유의미함.
```

### PCA

```python
scaled_variable_frame.shape
```

```python
from sklearn.decomposition import PCA

# PCA 수행
pca = PCA(n_components=25)
pca.fit(scaled_variable_frame)
variable_pca = pca.transform(scaled_variable_frame)
```

```python
print(np.sum(pca.explained_variance_ratio_))
```

```python
# 각 주성분의 설명된 분산 비율
explained_variance = pca.explained_variance_ratio_

# Scree Plot 그리기
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.xticks(range(1, len(explained_variance) + 1))
plt.grid()
plt.show()
```

```python
# 각 주성분의 설명된 분산 비율
explained_variance = pca.explained_variance_ratio_

# Scree Plot 그리기
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.xticks(range(1, len(explained_variance) + 1))
plt.grid()
plt.show()
```

```python
cumulative_variance = np.cumsum(explained_variance)
print(cumulative_variance)
# 누적 기여율이 80% 이상이 넘어가는 지점까지의 주성분을 기준으로 한다고 함.
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

# 2. 유사도 측정 및 계층적 클러스터링 수행
Z = linkage(variable_pca8, method='ward')  # 'ward'는 분산 최소화를 기준으로 함

# 3. 덴드로그램 시각화
plt.figure(figsize=(10, 7))
dendrogram(Z, labels = df.index)
plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')

for label in plt.gca().get_xticklabels():
    if label.get_text() in transfer:
        label.set_color('red')

plt.show()
```

고려해야 할 점<br>
- 해석 가능성: PCA로 생성된 주성분은 원래 변수의 조합이기 때문에, 결과의 해석이 어려울 수 있습니다. 원래 변수의 의미를 잃을 수 있다는 점을 유의해야 합니다.

- 모델의 목표: 만약 특정 독립 변수가 중요한 해석적 의미를 가지거나 정책적 결정을 위한 기준이 된다면, PCA를 사용하는 것이 적절하지 않을 수 있습니다.

- 데이터의 특성: 데이터가 이미 적은 차원을 가지고 있거나, 주성분이 종속 변수를 잘 설명하지 못할 경우 PCA의 이점이 적을 수 있습니다.

PCA는 특정 상황에서 OLS 회귀 분석의 성능을 향상시킬 수 있지만, 항상 최선의 선택은 아닙니다. 데이터의 특성과 분석 목표에 따라 PCA의 필요성을 신중하게 판단해야 합니다.

### VIF
VIF(Variance Inflation Factor)를 계산할 때는 일반적으로 스케일링된 변수를 사용하지 않고 원래 변수에 대해 계산하는 것이 좋다고 함.

```python
variable = variable.astype({col: 'float32' for col in variable.select_dtypes(include=['int64']).columns})
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from matplotlib import rc

# VIF 출력을 위한 데이터 프레임 형성
vif = pd.DataFrame()

# VIF 값과 각 Feature 이름에 대해 설정
vif["VIF Factor"] = [variance_inflation_factor(variable.values, i) for i in range(variable.shape[1])]
vif["features"] = variable.columns

# VIF 값이 높은 순으로 정렬
vif = vif.sort_values(by="VIF Factor", ascending=False)
vif = vif.reset_index().drop(columns='index')
vif
```
